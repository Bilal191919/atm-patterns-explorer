```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = 'atm_transactions.csv'
atm_data = pd.read_csv(file_path)
```

```python
atm_data.head(5)
```

```python
atm_data.tail(5)
```

```python
atm_data['atmName'] = atm_data['atmName'].replace({
    'ALİAĞA BAĞLI ŞB': 'Allied BRANCH',
    'ALİAĞA İSTİKLAL CADDESİ': 'Allied INDEPENDENCE STREET',
    'ALİAĞA PAZARYERİ': 'Allied MARKET',
    'ALİAĞA SAHİL PARKI': 'Allied BEACH PARK',
    'ALİAĞA ŞUBE-1': 'ALLied BRANCH-1'
})
atm_data['atmCity'] = atm_data['atmCity'].replace({"Izmir":"lahore"})

atm_data['atmAddress'] = atm_data['atmAddress'].replace({"Aliağa Petkim Tesisleri":	"Allied thokar branch",
                                                         "ATATÜRK İSTİKLAL CD. NO:97":"Awan town",
                                                         "ALİAĞA KAPALI PAZAR GİRİŞİ":"Samanabad",
                                                         "ALİAĞA HİKMET CD.SAHİL PARKI ÖNÜ":"Iqbal town"
                                                         ,"İstiklal Caddesi No: 64 / A":"Ali town"})

atm_data.head(5)
```

```python
print("Translated ATM Data:")
print(atm_data[['atmId', 'atmName', 'atmCity', 'totalBalance', 
                'numberIncomeTransaction', 'numberOutcomeTransaction', 'totalIncome', 
                'totalOutcome', 'totalNumberTransaction', 'day', 'transactionTime']])
```

```python

total_transactions = atm_data['totalNumberTransaction'].sum()
print(f"Total Number of Transactions: {total_transactions}")
atm_data['totalNumberTransaction'].value_counts().plot(kind="bar", color="Blue")
plt.show()
```

```python
totalBalance = atm_data['totalBalance'].sum()
print(f"totalBalance: {totalBalance}")
atm_data['totalBalance'].value_counts().plot(kind="line", color="red")
plt.show()
```

```python
atm_data.to_csv('final_result.csv', index=False)
```
